```python
df = pd.read_csv("data-export (1).csv", header=1)
df.head()
```

```python
df.info()
```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"],format="%Y%m%d%H",errors="coerce")
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.iloc[0]
df = df.drop(index=0).reset_index(drop=True)
```

```python
df["DateHour"] = pd.to_datetime(
    df["DateHour"],
    format="%Y%m%d%H",
    errors="coerce"
)
```

```python
df.head()
```

```python
numeric_cols = df.columns.drop(["channel group", "DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric,errors="coerce")
df["Hour"] = df["DateHour"].dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,6))
df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and Users Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data=df,
    x="channel group",
    y="Users",
    estimator=np.sum)
plt.title("Total Users by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data=df,
    x="channel group",
    y="Average engagement time per session",
    estimator=np.mean)
plt.title("Average Engagement Time by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(
    data=df,
    x="channel group",
    y="Engagement rate")
plt.title("Engagement Rate Distribution by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
session_df = df.groupby("channel group")[
    ["Sessions","Engaged Sessions"]].sum()
session_df["Non-Engaged"] = (
    session_df["Sessions"]
    - session_df["Engaged Sessions"])
session_df = session_df.reset_index()
session_df_melted = session_df.melt(
    id_vars="channel group",
    value_vars=["Engaged Sessions","Non-Engaged"])
plt.figure(figsize=(10,6))
sns.barplot(
    data=session_df_melted,
    x="channel group",
    y="value",
    hue="variable")
plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

```python
heatmap_data = df.groupby(
    ["Hour","channel group"]
)["Sessions"].sum().unstack()
plt.figure(figsize=(12,6))
sns.heatmap(
    heatmap_data,
    cmap="YlGnBu",
    annot=True,
    fmt=".0f")
plt.title("Traffic by Hour and Channel")
plt.xlabel("Channel")
plt.ylabel("Hour")
plt.show()
```

```python
df_plot = df.groupby("DateHour")[
  ["Engagement rate","Sessions"]
].mean().reset_index()
plt.figure(figsize=(12,6))
plt.plot(
    df_plot["DateHour"],
    df_plot["Engagement rate"],
    label="Engagement Rate")
plt.plot(
    df_plot["DateHour"],
    df_plot["Sessions"],
    label="Sessions")
plt.title("Engagement Rate vs Sessions Over Time")
plt.xlabel("Date")
plt.legend()
plt.grid(True)
plt.show()
```

```python
top_channel = df.groupby("channel group")["Users"].sum().idxmax()
print("Top Channel:", top_channel)
print("\nAverage Engagement Rate:")
print(
    df.groupby("channel group")["Engagement rate"]
    .mean()
    .sort_values(ascending=False))
print("\nAverage Engagement Time:")
print(
    df.groupby("channel group")
    ["Average engagement time per session"]
    .mean()
    .sort_values(ascending=False)
)
```
